## Background:

Formula 1, often abbreviated as F1, is known as the pinnacle of motorsport racing. It is one of world's most prestigious international single-seater auto racing. Formula 1 world championship is organized by Federation International Automobile (FIA), which consist of multiple races called Grand Prix, held on various circuits around the world. Formula 1 came from the early 20th century, with the first World Championship taking place in 1950. 

The championship was established by the FIA, which developed various Grand Prix motor racing events into a single championshp. The championship is won by teams and each with two drivers. These technical teams design, develop, test, and race their built cars, which are subject to strict regulations and budget cap by FIA. Qualifying sessions in F1 determines the starting positions for each car on the grid for race day. During qualifying, setting the fastest lap time determines the starting position for the grid, so it is crucial as it sets the stage for the race and significantly impact driver's success of being a Grand Prix winner on race day. Therefore, this model will provide whether qualifying time and race winners have a significant impact.

## Exploratory data analysis (EDA)

### Setting up our data source for analysis

```python
#Joining all dataframe originally called into one main dataframe as df_combined
df_merge1 = pd.merge(df_results, df_races, on ='raceId')
df_merge2 = pd.merge(df_merge1, df_qual, on ='driverId')
df_merge3 = pd.merge(df_merge2, df_drivers, on ='driverId')
df_merge4 = pd.merge(df_merge3, df_circuit, on ='circuitId')
race_df = pd.merge(df_merge4, df_status, on ='statusId')
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
race_df.sample(5)
```

```python
#Dropping columns and including the columns only for our model 
race_df = race_df[['resultId', 'raceId_x', 'grid', 'position_x', 'year', 'name_x', 'name_y',
                   'q1', 'q2', 'q3', 'forename', 'surname', 'country', 'status']]
race_df.head()
```

```python
#Adding forename and surname together into a single column 
race_df['driver name'] = race_df['forename'] + ' ' + race_df['surname']
race_df.sample(5)
```

```python
#Displaying the new column for driver name and dropping forename and surname from dataframe
drop_columns = ['forename', 'surname']
race_df = race_df.drop(columns = drop_columns)
race_df.head() #Now we can see that the driver names are combined, forename and surname are dropped
```

```python
#Checking for null values 
race_df.isnull().sum()
```

## Handling null values 
From our analysis, we can see that there are many null values in the columns q1, q2, and q3. However, this is expected because many drivers during qualifying session do not make it all the way through each qualifying. So instead, we will replacing the null values with string value as 'DNF' (did not finish) as it's a term in F1. 

```python
#Changing the qualifying columns to DNF where they are null
replace_columns = ['q1', 'q2', 'q3']

#Replacing the race_df columns ('q1', 'q2', 'q3') with 'DNF'
race_df[replace_columns] = race_df[replace_columns].replace(r'\N', 'DNF')
#replace_value = 'DNF'

#race_df[replace_columns] = race_df[replace_columns].fillna(replace_value)
```

```python
#Displaying the information of the dataset race_df
race_df.info()
```

```python
#Check for null values again 

#null_values = race_df.isnull()
#print(null_values)

race_df.isnull().sum() #We can see that the null values are gone
```

We can see that the null values are still here. However, we will replaced all the qualifying session to 'DNF' (Did not finish).

```python
# Dataframe to show rows with 'DNF' values
df_dnf = race_df[race_df.eq('DNF').any(axis=1)]

# Display the 'DNF' values in the dataframe 
df_dnf.sample(5)
```

Notice that there are additonal '\N' values for the position_x column. We will replace these values as 'DNF' (Did not finish) because they did not finish the race. We will be replacing the column position_x with 'DNF' where they show '\N'. 

```python
#Changing the qualifying columns to DNF where they are null
replace_columns2 = ['position_x']


race_df[replace_columns2] = race_df[replace_columns2].replace(r'\N', 'DNF')
```

```python
# Specify the column name
column_name2 = 'position_x'

# Filter the dataframe to show only rows with 'DNF' in the specified column
dnf_rows2 = race_df[race_df[column_name2] == 'DNF']

# Display the resulting dataframe
dnf_rows2.sample(5)
```

## Handling naming conventions for columns and dataframe

```python
#Renaming the columns 
#Define dictionary and map them current column names to new column names 
new_columns = {'resultId': 'result_id', 
             'raceId_x': 'race_id', 
             'grid': 'starting_position', 
             'position_x': 'final_position', 
             'year': 'year', 
             'name_x': 'grand_prix', 
             'name_y': 'circuit_name', 
             'driver name': 'driver_name'}

race_df = race_df.rename(columns=new_columns) #Renaming the columns 
column_names = race_df.columns.tolist()
print(column_names) #Printing all the column names (updated)
```

```python
#Displaying all the data with the new naming convention
race_df.head(5)
```

```python
#Reordering the dataframe to follow specific column
desired_order = ['result_id', 'race_id', 'year', 
                  'grand_prix', 'circuit_name', 'country', 'q1', 'q2', 'q3',
                  'driver_name', 'starting_position', 'final_position', 'status']

race_df = race_df.reindex(columns = desired_order) #reindexing the colmuns of the dataframe
race_df.sample(5)
```
